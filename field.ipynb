```python
%matplotlib inline
from scipy.io import *
import matplotlib.pyplot as plt
import numpy as np
import pyfield
from sklearn.cluster import KMeans
```

```python
s = np.load('data\sdo.x766y1118.20110225.npz')

data = np.sum(s['hmi'][144:147,:,:],0)/3.
data[np.where(np.abs(data) < 50)] = 0
```

```python
field = pyfield.pfield(data)
nlines = 100

lines = []
pos = np.empty((nlines,2))
neg = np.empty((nlines,2))


for i in range(0,nlines):
    x,y = np.random.normal(32, 8, size = 2).clip(min=0, max=64)
    line = pyfield.fline(field, x,y,64,64)
    lines.append(line)
    
    pos[i,:] = line[:,-1]
    neg[i,:] = line[:,0]
    
```

```python
k_neg = KMeans(n_clusters = 3)
idx = np.where(np.max(np.abs(neg-32), axis = 1) < 25)
dom = k_neg.fit_predict(neg[idx])

dom_neg = np.zeros(nlines, dtype = np.int)-1
dom_neg[idx] = dom

####################

k_pos = KMeans(n_clusters = 4)
idx = np.where(np.max(np.abs(pos-32), axis = 1) < 25)
dom = k_pos.fit_predict(pos[idx])

dom_pos = np.zeros(nlines, dtype = np.int)-1
dom_pos[idx] = dom

####################

domains = (dom_pos, dom_neg)
```

```python
colors = ['r','g','b','c','m','brown','k','tan','dimgrey','darkblue',
          'darkred','lime']
fig = plt.figure(figsize=(14, 14))
plt.imshow(data)

for k in range(0,4):
    for j in range(0,3):
        ids = np.where(np.all([domains[0] == k, domains[1] == j], axis=0))[0]


        for i in ids:
            plt.plot(lines[i][0,:], lines[i][1,:], color=colors[j*4+k])
```
