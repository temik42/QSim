```python
%matplotlib inline
from scipy.io import *
import matplotlib.pyplot as plt
import numpy as np
from pyfield import *
import time
```

```python
s = np.load('data\sdo.x766y1118.20110225.npz')

hmi_data = s['hmi']
aia_data = s['aia_171']
field1 = Field(hmi_data[127,:,:],1.).search_charges(25).search_nulls()
field2 = Field(hmi_data[135,:,:],1.).search_charges(25).search_nulls()
```

```python
fig = plt.figure(figsize = (16,10))
plt.subplot(1,2,1)
field1.draw_footprint(fig)
plt.xlabel('before the flare', size = 18)
plt.subplot(1,2,2)
field2.draw_footprint(fig)
plt.xlabel('after the flare', size = 18)
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.imshow(np.flipud(aia_data[500,:,:])**0.3, 'gray', interpolation = 'nearest')
```

```python
plt.figure(figsize=(16,8))
for j in range(0,4):
    for i in range(0,8):
        plt.subplot(4,8,i+j*8+1)
        plt.imshow(np.flipud(aia_data[480+(i+8*j)*5,:,:])**0.3, 'gray', interpolation = 'nearest')
```

```python
plt.figure(figsize=(10,7))
plt.title('AIA 171 intensity at the core of the flare', size = 20)
plt.plot(np.arange(0,1800)*12./60, np.log10(aia_data[:,28,34]))
plt.axis([0,350,1.5,4])
plt.xlabel('Time, min', size = 18)
plt.ylabel('Intensity, counts (log scale)', size = 18)
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.title('Magnetic field strength of the negative source', size = 20)
plt.plot(np.arange(0,481)*45./60, hmi_data[:,19,49]*0.1)
plt.axis([0,350,-70,10])
plt.xlabel('Time, min', size = 18)
plt.ylabel('Bz, Gauss', size = 18)
plt.show()
```
