```python
%matplotlib inline
import sys
sys.path.append('Q:\\python\\lib')
from scipy.io import *
from scipy import misc
import matplotlib.pyplot as plt
import numpy as np
from pyfield import *
import time
```

```python
s = np.load('sdo.x766y1118.20110225.npz')
hmi_data = s['hmi']
aia_data = s['aia_171']
aia_193 = s['aia_193']
aia_131 = s['aia_131']
```

```python
plt.figure(figsize=(10,7))
plt.title('AIA 171 intensity at the core of the flare', size = 20)
plt.plot(np.arange(0,1800)*12./60, aia_data[:,28,34])
plt.axis([0,350,0,1e4])
plt.xlabel('Time, min', size = 18)
plt.ylabel('Intensity, counts', size = 18)
plt.show()
```

```python
plt.figure(figsize=(16,8))
for j in range(0,4):
    for i in range(0,8):
        plt.subplot(4,8,i+j*8+1)
        plt.imshow(np.flipud(aia_data[480+(i+8*j)*5,:,:])**0.3, 'gray', interpolation = 'nearest')
```

```python
field = Field(hmi_data[100,:,:],1.).search_charges(25).search_nulls()
```

```python
fig = plt.figure(figsize = (16,10))
plt.subplot(1,2,1)
plt.title('HMI photospheric magnetic field', size = 20)
plt.imshow(np.flipud(hmi_data[100,:,:]), 'gray', interpolation = 'nearest')
plt.subplot(1,2,2)
plt.title('Topological skeleton', size = 20)
field.draw_footprint(fig)
plt.show()
```
