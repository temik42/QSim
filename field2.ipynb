```python
%matplotlib inline
from scipy.io import *
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import numpy.fft as fft
import scipy.interpolate as interp
import scipy.integrate as integr
import pyfield
import pymc as pm
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
import scipy.optimize
```

```python
s = np.load('data\sdo.x766y1118.20110225.npz')

data = np.sum(s['hmi'][144:147,:,:],0)/3.
data[np.where(np.abs(data) < 10)] = 0
```

```python
class Field:
    def __init__(self, data):
        self.set_phField(data)
            
        
    def set_phField(self, data):
        self.dim = np.shape(data)

        temp = np.pad(data,((0,self.dim[0]),(0,self.dim[1])),'constant')
    
        xi = np.roll(np.arange(-self.dim[0],self.dim[0]),self.dim[0])
        yi = np.roll(np.arange(-self.dim[1],self.dim[1]),self.dim[1])
    
        xi = np.resize(xi,(2*self.dim[0],2*self.dim[1]))/(2.*self.dim[0])
        yi = (np.resize(yi,(2*self.dim[1],2*self.dim[0]))/(2.*self.dim[1])).T
        
        q = np.sqrt(xi**2 + yi**2)
        fdata = np.fft.fft2(temp)*np.exp(-2*np.pi*q)
    
        gx = xi*(-1j)/q.clip(min = (1./(self.dim[0]*self.dim[1])))   
        gy = yi*(-1j)/q.clip(min = (1./(self.dim[0]*self.dim[1])))   
    
        bx = np.real(np.fft.ifft2(fdata*gx))[0:self.dim[0],0:self.dim[1]]
        by = np.real(np.fft.ifft2(fdata*gy))[0:self.dim[0],0:self.dim[1]]
        bz = np.real(np.fft.ifft2(fdata))[0:self.dim[0],0:self.dim[1]]
        
        self.phField = (bx, by, bz)
        
        bx_interp = interp.interp2d(np.arange(self.dim[0]),np.arange(self.dim[1]),bx)
        by_interp = interp.interp2d(np.arange(self.dim[0]),np.arange(self.dim[1]),by)
        bz_interp = interp.interp2d(np.arange(self.dim[0]),np.arange(self.dim[1]),bz)
        
        self.phFieldInterpolator = (bx_interp, by_interp, bz_interp)
        return self
     
    def get_phField(self, x):
        return (self.phFieldInterpolator[0](x[0],x[1])[0], 
                self.phFieldInterpolator[1](x[0],x[1])[0], 
                self.phFieldInterpolator[2](x[0],x[1])[0])
    
    def search_charges(self, maxi = 1000):
        x = np.empty(2)
        
        for i in range(0,maxi):     
            x0 = (np.random.normal(self.dim[0]/2, self.dim[0]/4),np.random.normal(self.dim[1]/2,self.dim[1]/4))
            xmin = scipy.optimize.fmin(lambda x: - np.abs(self.phFieldInterpolator[2](x[0],x[1])[0]), 
                                 x0, disp = False)
            
            if (abs(xmin[0]-self.dim[0]/2) < self.dim[0]/2 and 
                abs(xmin[1]-self.dim[1]/2) < self.dim[1]/2 and
                np.abs(self.phFieldInterpolator[2](xmin[0],xmin[1])[0]) > 20):
                x = np.column_stack((x,xmin))
        
        db = DBSCAN()
        db.fit_predict(x.T)
        
        self.n_charges = np.max(db.labels_)
        
        self.xi = np.zeros((3,self.n_charges))
        self.qi = np.zeros(self.n_charges)
        signs = np.zeros(self.n_charges)
        
        for i in range(0, self.n_charges):
            self.xi[0:2,i] = np.mean(x[:,db.labels_ == i], axis=1)
            self.qi[i] = self.phFieldInterpolator[2](self.xi[0,i],self.xi[1,i])[0]
        
        #self.qi /= np.max(np.abs(self.qi))
        return self
        
        signs = np.sign(self.qi)
        
        charges = pm.Uniform('charges', 0, 1, value = np.abs(self.qi), size = self.n_charges)

        @pm.deterministic(dtype=float)
        def sim(charges = charges): 
            self.qi = charges*signs
            return self.allign()

        @pm.stochastic(observed = True)
        def prob(sim=sim, value = 1):
            def logp(sim, value):
                return sim
        
        
        
        model = pm.Model([charges, sim, prob])
        mcmc = pm.MCMC(model)
        mcmc.sample(2000,1000,10)
        
        return self
    
    def get(self, x):
        self._x = x
        self.dx = (x - self.xi.T).T
        self.dr = (np.sum(self.dx**2, axis = 0))**1.5
        self.argminr = np.argmin(self.dr)
        result = np.sum(self.dx*self.qi/self.dr, axis = 1)
        return result
    
    def rk2(self, x, mul = 1, maxi = 100, step = 0.5):
        self._x = x
        result = self._x
        
        i = 0
        while self._x[2] > 0 and i < maxi:
            
            p = self.get(self._x)
            dx = p/np.sqrt(np.sum(p**2))*step*mul
            p = 0.5*(p+self.get(self._x+dx))
            dx = p/np.sqrt(np.sum(p**2))*step*mul
            
            self._x += dx
            if self._x[2] < 0:
                self._x = self.xi[:,self.argminr]
            result = np.column_stack((result,self._x))
            i += 1
    
        return result
    
    def allign(self, maxi = 1000):
        xi = np.column_stack((np.random.normal(self.dim[0]/2,self.dim[0]/4,maxi),
                              np.random.normal(self.dim[1]/2,self.dim[1]/4,maxi))).T
        result = 0
        
        for i in range(0,maxi):
            pf = np.array(self.get_phField(xi[:,i]))
            f = self.get(np.append(xi[:,i],1))
            result += np.dot(f,pf)/(np.linalg.norm(f)*np.linalg.norm(pf))/maxi
            
        return result
    
        
    def fline(self, x):
        return np.column_stack((np.fliplr(self.rk2(x)), self.rk2(x, mul = -1)))


field = Field(data).search_charges()
field.allign()
```
